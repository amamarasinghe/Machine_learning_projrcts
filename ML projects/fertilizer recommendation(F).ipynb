```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/amama/Downloads/Crop_recommendation.csv")
```

```python
print("Rows: ", df.shape[0])
print("Columns: ", df.shape[1])
```

```python
# Get the first five rows of the data
print(df.head())
```

```python
# Get the summary statistics of the data
print(df.describe())
```

```python
# Check for missing values in the data
print(df.isnull().sum())
# No missing values found
```

```python
# Dropping column 'Label' since it's not numeric
exclude_column = 'label'
df = df.drop(exclude_column, axis=1)
```

```python
# Plotting the heatmap
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# Plot boxplot
sns.boxplot(data=df)
plt.show()
```

```python
# Plot histograms
df.hist(bins=30, figsize=(20, 15))
plt.show()
```

```python
# Plot count plots for int columns
for column in df.columns:
    if df[column].dtype == 'int64':
        sns.countplot(x=column, data=df)
        # plt.show()
```

```python
# Plot count plots for float columns
for column in df.columns:
    if df[column].dtype == 'float64':
        sns.countplot(x=column, data=df)
        # plt.show()
```

```python
# Plot a scatter plot matrix
x_column = 'humidity'
y_column = 'temperature'
a_column = 'N'
b_column = 'P'
c_column = 'K'
d_column = 'ph'
e_column = 'rainfall'

figsize = (3, 2)
```
